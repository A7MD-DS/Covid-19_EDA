```python
import pandas as pd 

df = pd.read_csv('../data/country_wise_latest.csv')
```

```python
print(df.head)
print(df.info)
print(df.describe)
```

```python
print(df.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#group data
grouped = df.groupby('Country/Region')['Confirmed'].sum().sort_values(ascending=False).head(20)

#plot
plt.figure(figsize=(10, 6))
sns.barplot(x=grouped.values, y=grouped.index, palette='rocket')
plt.title("Top 20 Countries by Confirmed COVID-19 Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
#group data
grouped = df.groupby('Country/Region')['Deaths / 100 Cases'].sum().sort_values(ascending=False).head(20)

#plot
plt.figure(figsize=(10, 6))
sns.barplot(x=grouped.values, y=grouped.index, palette='rocket')
plt.title("Top 20 Countries by Confirmed Deaths in every 100 COVID-19 Cases")
plt.xlabel("Confirmed Deaths")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
#group data
grouped = df.groupby('Country/Region')['Recovered / 100 Cases'].sum().sort_values(ascending=False).head(20)

#plot
plt.figure(figsize=(10, 6))
sns.barplot(x=grouped.values, y=grouped.index, palette='rocket')
plt.title("Top 20 Countries by Confirmed Recovered in every 100 COVID-19 Cases")
plt.xlabel("Confirmed Recovered")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```
